```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('aanumerical.csv') 
df.head()
```

# Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.


```python

# Loop through each column and create distribution plots
for column in df.columns:
    plt.figure(figsize=(8, 6))
    sns.distplot(df[column], hist=True, kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
# Loop through each column and create distribution plots
for column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.distplot(df[column], hist=True, kde=True, bins=30, color='blue', fit=stats.norm)  # Fit to normal distribution
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    
    # Perform goodness-of-fit test for normal distribution
    _, p_value = stats.normaltest(df[column])
    # If the P-Value of the KS Test is larger than 0.05, we assume a normal distribution.
    #If the P-Value of the KS Test is smaller than 0.05, we do not assume a normal distribution
    if p_value < 0.05:
        plt.text(0.7, 0.9, "Not Normal", transform=plt.gca().transAxes, color='red')
    else:
        plt.text(0.7, 0.9, "Normal", transform=plt.gca().transAxes, color='green')
    
    plt.show()
```

# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python

from scipy import stats

# List of transformation functions to try
transformation_functions = [
    ('Original', lambda x: x),
    ('Log-transform', np.log1p),
    ('Square root transform', np.sqrt),
    ('Box-Cox transform', stats.boxcox)
]

# Loop through each column and apply different transformations
for column in df.columns:
    plt.figure(figsize=(12, 8))
    
    for label, transformation in transformation_functions:
        transformed_data = transformation(df[column])
        
        plt.subplot(len(transformation_functions), 1, transformation_functions.index((label, transformation)) + 1)
        sns.distplot(transformed_data, hist=True, kde=True, bins=30, color='blue', label=label)
        plt.title(f'Distribution of {column} ({label})')
        plt.xlabel(column)
        plt.ylabel('Density')
        plt.legend()
        
        _, p_value = stats.normaltest(transformed_data)
        if p_value < 0.05:
            plt.text(0.7, 0.9, "Not Normal", transform=plt.gca().transAxes, color='red')
        else:
            plt.text(0.7, 0.9, "Normal", transform=plt.gca().transAxes, color='green')
    
    plt.tight_layout()
    plt.show()
```
